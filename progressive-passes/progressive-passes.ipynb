# Progressive Passes

A pass that gets the ball closer to the goal

### American Soccer Analysis Definition
Our definition of progressive passes at American Soccer Analysis is a bit different than other variations on the theme: all are in the attacking 60% of the pitch and must travel 25% closer to the goal. Some of these progressive passes are used ahead of a key pass, but some are very specifically used to create a high-quality shot.
https://www.americansocceranalysis.com/home/2021/2/24/where-goals-come-from-passing-in-the-final-third#:~:text=Our%20definition%20of%20progressive%20passes%20at%20American%20Soccer,very%20specifically%20used%20to%20create%20a%20high-quality%20shot.

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from mplsoccer import Pitch
```

```python
df = pd.read_csv('europaFinal.csv')
```

```python
df
```

```python
df = df.loc[df['teamId']=='Villarreal']
```

```python
df = df.loc[df['type']=='Pass']
```

```python
df = df.loc[df['playerId']==5.0].reset_index()
```

```python
df = df.loc[df['progressive']==True].reset_index()
```

```python
df.x = df.x*1.2
df.y = df.y*.8
df.endX = df.endX*1.2
df.endY = df.endY*1.2
```

```python
df['beginning'] = np.sqrt(np.square(120 - df['x']) + np.square(40 - df['y']))
df['end'] = np.sqrt(np.square(120 - df['endX']) + np.square(40 - df['endY']))
```

```python
df['progressive'] = [(df['end'][x]) / (df['beginning'][x]) < .75 for x in range(len(df.beginning))]
```

```python
pitch = Pitch(pitch_type='statsbomb')
fig, ax = pitch.draw()
plt.title('Dani Parejo Progressive Passes Europa League Final')

pitch.lines(df.x, df.y, df.endX, df.endY, comet=True, ax=ax);
```
